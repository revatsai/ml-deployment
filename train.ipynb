```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
import pandas as pd
import pickle as pkl
```

```python
df = pd.read_csv('./data/training_data.csv')
df.head()
```

```python
df['category'].unique() # 需要轉成數字的類型
```

```python
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(df['category'])
df['label'] = label_encoder.transform(df['category']) # 進行數值的轉換
```

```python
df['label'].unique()
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
```

```python
x = df['body']
y = df['label']
```

```python
vectorized_x = vectorizer.fit_transform(x)
vectorized_x
```

```python
rf_clf = RandomForestClassifier()
rf_clf.fit(vectorized_x, y)
```

```python
pkl.dump(rf_clf, open('./models/text_classifier.pkl', 'wb'))
pkl.dump(vectorizer, open('./models/text_vectorizer.pkl', 'wb'))
pkl.dump(label_encoder, open('./models/text_encoder.pkl', 'wb'))
```
